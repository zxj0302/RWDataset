```python
import subprocess
import ipaddress
import paramiko

def scan_network(base_ip):
    network = ipaddress.IPv4Network(f"{base_ip}.0/24", strict=False)
    for ip in network.hosts():
        yield str(ip)

def try_ssh_connection(ip, username, port=22, timeout=5):
    ssh = paramiko.SSHClient()
    ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    try:
        ssh.connect(ip, port=port, username=username, timeout=timeout)
        return True
    except (paramiko.AuthenticationException, paramiko.SSHException):
        # Authentication failed, but the server exists
        return True
    except:
        return False
    finally:
        ssh.close()

def find_server():
    base_ip = "10.64.140"
    username = "zxj"
    port = 22

    for ip in scan_network(base_ip):
        print(f"Trying {ip}...")
        if try_ssh_connection(ip, username, port):
            print(f"Server found! IP: {ip}")
            return ip

    print("Server not found.")
    return None

if __name__ == "__main__":
    server_ip = find_server()
    if server_ip:
        print(f"You can connect to your server using: {server_ip}")
```
